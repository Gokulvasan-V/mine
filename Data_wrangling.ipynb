# **Data wrangling**

```python
import pandas as pd
```

```python
# import data directly from website
link="https://raw.githubusercontent.com/chrisalbon/simulated_datasets/master/titanic.csv"
data_set= pd.read_csv(link)
# showing 1st 5 data's
data_set.head()
```

**creating our own Dataframe**

```python
# here we created emty Dataframe
data_frame=pd.DataFrame()
# here we are adding colomns
data_frame['Names']=['Gokul','Naveen','Muthuraj']
data_frame['Ages']=[22,22,23]
data_frame['salary(in thousands)']=[30,25,15]
# printing our created Dataframe
data_frame
```

```python
# adding new row
new_row=pd.Series(['Mani',22,51],index=['Names','Ages','salary(in thousands)'])
data_frame.append(new_row,ignore_index=True)
```
